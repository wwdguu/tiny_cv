# Implemention of Histogram Equalization Algorithms

```python
## load images and compute histogram
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

src=cv2.imread('haze_img/12.jpg')
plt.title('src')
plt.axis('off')
plt.imshow(src[:,:,::-1])
```

```python
gray=cv2.cvtColor(src,cv2.COLOR_RGB2GRAY)
plt.title('gray')
plt.axis('off')
plt.imshow(gray,cmap='gray')
```

```python
def hist(img_single_channel):
    hist=np.bincount(img_single_channel.reshape(-1,))
    hist=list(hist)+(256-len(hist))*[0]
    return np.array(hist)
h_gray=hist(gray)
h_blue=hist(src[:,:,0])
h_green=hist(src[:,:,1])
h_red=hist(src[:,:,2])
plt.plot(range(len(h_gray)),h_gray,label='gray',color='k')
plt.plot(range(len(h_blue)),h_blue,label='blue',color='b')
plt.plot(range(len(h_green)),h_green,label='green',color='g')
plt.plot(range(len(h_red)),h_red,label='red',color='r')
plt.legend()
plt.show()
```

compute cumulative distribution c(I)
$c(I)=\frac{1}{N}\sum_{i=0}^Ih(i)=c(I-1)+\frac{1}{N}h(I)$

```python

def compute_c(h):
    c=np.zeros((256,))
    N=np.sum(h)
    c[0]=h[0]*1./N
    for i in range(1,256):
        c[i]=c[i-1]+1.*h[i]/N
    return np.round((c*255).astype(np.uint8))
c_gray=compute_c(h_gray)
plt.plot(range(len(c_gray)),c_gray,label='c_gray',color='k')
plt.legend()
plt.show()
```

```python
# equalize hist using c
def base_equalize_hist(src_single_channel):
    h=hist(src_single_channel)
    c=compute_c(h)
    dst=np.ones_like(src_single_channel)
    h,w=dst.shape[:2]
    for i in range(h):
        for j in range(w):
            dst[i,j]=c[src_single_channel[i,j]]
    return dst

equalized=base_equalize_hist(gray)
plt.title('equalized gray')
plt.axis('off')
plt.imshow(equalized,cmap='gray')

```

```python
h_equalized=np.bincount(equalized.reshape(-1,))
plt.plot(range(len(h_equalized)),h_equalized,label='equalized',color='b')
plt.legend()
plt.show()
```

Using a mapping function
$f(I)=\alpha c(I)+(1-\alpha)I$

```python
def equalize_hist_with_alpha(src_single_channel,alpha=0.5):
    h=hist(src_single_channel)
    c=compute_c(h)
    dst=np.ones_like(src_single_channel)
    h,w=dst.shape[:2]
    for i in range(h):
        for j in range(w):
            dst[i,j]=alpha*c[src_single_channel[i,j]]+(1-alpha)*src_single_channel[i,j]
    return dst.astype(np.uint8)

equalized_with_alpha=equalize_hist_with_alpha(gray,alpha=0.8)
plt.title('equalized gray with alpha')
plt.axis('off')
plt.imshow(equalized_with_alpha,cmap='gray')
```

```python
h_equalized_with_alpha=hist(equalized_with_alpha)
plt.plot(range(len(h_equalized_with_alpha)),h_equalized_with_alpha,label='equalized_with_alpha',color='b')
plt.legend()
plt.show()
```

Equalize on all channels

```python
b,g,r=cv2.split(src)
Y=gray
Y_equalized=base_equalize_hist(np.round(Y).astype(np.uint8))
sum_=np.sum(src,axis=2)
eps=1e-5
b_ratio=b*1./(sum_+eps)
g_ratio=g*1./(sum_+eps)
r_ratio=r*1./(sum_+eps)
Y_ratio=Y_equalized/(Y+eps)
b_new=Y_ratio*b_ratio
g_new=Y_ratio*g_ratio
r_new=Y_ratio*r_ratio

color_equalized=(cv2.merge((b_new,g_new,r_new))*255).astype(np.uint8)
plt.imshow(color_equalized[:,:,::-1])
```

```python
h_gray=hist(Y)
h_blue=hist(color_equalized[:,:,0])
h_green=hist(color_equalized[:,:,1])
h_red=hist(color_equalized[:,:,2])
plt.plot(range(len(h_gray)),h_gray,label='gray',color='k')
plt.plot(range(len(h_blue)),h_blue,label='blue',color='b')
plt.plot(range(len(h_green)),h_green,label='green',color='g')
plt.plot(range(len(h_red)),h_red,label='red',color='r')
plt.legend()
plt.show()
```
